Part D: Principal Component Analysis - PCA (Dimensionality Reduction)

```python
# Subset numeric data only
numeric_columns = ['Salary', 'EngagementSurvey', 'EmpSatisfaction',
       'SpecialProjectsCount', 'DaysLateLast30',
       'Absences', 'Age']
df = hr[numeric_columns].head()
```

```python
# Scaling the data
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
```

```python
# Standardize numerical features before applying PCA.
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
```

```python
# Apply PCA
pca = PCA(n_components=7)
pca_result = pca.fit_transform(scaled)
#Plot results
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Result")
plt.grid(True)
plt.show()
```

```python
# Plot the explained variance ratio for the PCA components.
# Choosing the numbe of components (scree plot)
pca = PCA().fit(scaled)
plt.plot(range(1, len(pca.explained_variance_ratio_)+1),
    pca.explained_variance_ratio_.cumsum(), marker = 'o')
plt.xlabel('Number of components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.title('Explained variance vs. Number of components')
plt.show()

```

```python
# separate features and labels
le = LabelEncoder()
hr['Department_encoded'] = le.fit_transform(hr['Department'])
X = hr[numeric_columns]
y = hr['Department_encoded']
```

```python
# Standardize features before PCA
X_scaled = StandardScaler().fit_transform(X)
# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
pca_df['Department_encoded'] = y.values
```

```python
# Plot the PCA results colored by department
plt.figure(figsize=(8,6))
for dept in pca_df['Department_encoded'].unique():
    subset = pca_df[pca_df['Department_encoded'] == dept]
    plt.scatter(subset['PC1'], subset['PC2'], label=dept, alpha=0.7)

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Employees Colored by Department')
plt.legend()
plt.show()
```

```python
X = hr[['EngagementSurvey', 'EmpSatisfaction', 'Absences']]

# Standardize the variables
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA()
pca.fit(X_scaled)

# Explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Cumulative variance explained:", pca.explained_variance_ratio_.cumsum())

# Loadings (coefficients for each variable in components)
loadings = pd.DataFrame(pca.components_.T,
                        columns=[f'PC{i+1}' for i in range(len(pca.components_))],
                        index=X.columns)
print("\nLoadings:\n", loadings)
```

```python
# Drop non-numeric or target columns
X = hr[numeric_columns]
y = hr['PerformanceScore']

# Standardize before PCA
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA â†’ reduce to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Build PCA DataFrame
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
pca_df['PerformanceScore'] = y.values

# Plot with seaborn
plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='PerformanceScore', palette='Set2', alpha=0.8, s=80)

plt.title('Employees in PCA-Reduced 2D Space by Performance Score')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Performance Score')
plt.show()
```

```python

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# define features
X = hr[numeric_columns]
# Standardize features
X_scaled = StandardScaler().fit_transform(X)

# KMeans on original scaled data
kmeans_orig = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_orig = kmeans_orig.fit_predict(X_scaled)
silhouette_orig = silhouette_score(X_scaled, labels_orig)

print(f"Silhouette Score (original data): {silhouette_orig:.3f}")

# PCA to 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# KMeans on PCA-reduced data
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10)
labels_pca = kmeans_pca.fit_predict(X_pca)
silhouette_pca = silhouette_score(X_pca, labels_pca)

print(f"Silhouette Score (after PCA): {silhouette_pca:.3f}")

# Visualize clustering results after PCA
pca_df = pd.DataFrame(X_pca, columns=['PC1','PC2'])
pca_df['Cluster'] = labels_pca

plt.figure(figsize=(8,6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='Cluster', palette='Set2', s=80)
plt.title('KMeans Clusters After PCA')
plt.show()
```

Clustering after PCA shows much better separation, with the silhouette score improving from 0.316 to 0.636. This means PCA effectively captured the underlying structure, making the clusters more distinct and well-formed.

```python
X = hr[['Salary', 'Absences', 'EngagementSurvey']]

# Standardize features
X_scaled = StandardScaler().fit_transform(X)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Create DataFrame of PCA results
pca_df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])

# Get loadings (coefficients of variables on PCs)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Plot PCA points
plt.figure(figsize=(8,6))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.6)

# Plot loadings as arrows
for i, var in enumerate(['Salary','Absences','EngagementSurvey']):
    plt.arrow(0, 0, loadings[i,0]*2, loadings[i,1]*2, 
              color='red', alpha=0.7, head_width=0.05)
    plt.text(loadings[i,0]*2.2, loadings[i,1]*2.2, var, color='red', ha='center')

plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Biplot: Salary, Absences, EngagementSurvey')
plt.grid(True)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.show()
```

PCA is useful in HR analytics when you have many correlated variables and want to condense them into fewer, interpretable dimensions without losing much information. A common example is employee surveys: instead of analyzing 30â€“40 individual Likert-scale questions, PCA can reduce them into broader factors like job satisfaction, engagement, or workplace culture. This not only simplifies reporting but also reduces noise and multicollinearity in predictive models.

Itâ€™s also appropriate for exploratory analysis when the goal is to identify hidden patterns, such as grouping related performance indicators, training outcomes, or well-being measures. However, PCA is most effective when variables are continuous, moderately correlated, and measured on similar scalesâ€”otherwise the reduction may not be meaningful.

In HR contexts, PCA works best to summarize complex datasets (e.g., surveys, performance metrics, engagement indices) into a smaller set of interpretable dimensions that can then be used for clustering employees, predicting turnover, or tracking cultural shifts over time.
